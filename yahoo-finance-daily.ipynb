```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from selenium import webdriver

import sys

from selenium.webdriver.common.keys import Keys
driver = webdriver.Firefox()
import unittest, time, re
screener = pd.read_csv('nasdaq_screener_1615152187369.csv')

listOfStockNames = []
listOfStocks = []

for i in range(4,6):   #Load in each symbol one at a time, rip html from webpage
    val = screener.at[i,'Symbol']
    url = "https://ca.finance.yahoo.com/quote/" + val + "/history?period1=1457481600&period2=1615161600&interval=1d&filter=history&frequency=1d&includeAdjustedClose=true"
    driver.get(url)
    SCROLL_PAUSE_TIME = 0.3
    html = driver.find_element_by_tag_name('html')
    for j in range(0,15):  #Scroll through the contents of the page
        html.send_keys(Keys.END)
        time.sleep(SCROLL_PAUSE_TIME)

    page_source=driver.page_source
    pageRaw = BeautifulSoup(page_source, 'html.parser')  #Parse the data
    results = pageRaw.find_all('tr', attrs={'class', 'BdT Bdc($seperatorColor) Ta(end) Fz(s) Whs(nw)'})
    for result in results:  #Results refers to each table row
        rowValue = []
        skipDiv = False  #We don't want to store dividends
        for val in result.find_all('span'):
            if (val.text == "Dividend"):
                print("Skipped a Dividend")
                skipDiv = True
                break
            rowValue.append(val.text)
        if(skipDiv):
            continue
        else:
            stockStats.append(rowValue)
    print(len(stockStats))
    if(len(stockStats) < (1200 * 7)):
        continue
    else:
        df = pd.DataFrame(stockStats, columns = ['Date', 'Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume'])
        listOfStocks.append(df)
print(len(listOfStocks))
print(listOfStocks)
#result = driver.find_elements_by_xpath("//*[@class='BdT Bdc($seperatorColor) Ta(end) Fz(s) Whs(nw)']")
```

```python
listOfStockNames = []
listOfStocks = []
for i in range(5,9):  #len(screener):
    stockStats = []
    val = screener.at[i,'Symbol']
    listOfStocks.append(val)
    i += 1
    url = "https://ca.finance.yahoo.com/quote/" + val + "/history"
    r = requests.get(url)
    pageRaw = BeautifulSoup(r.text, 'html.parser')
    results = pageRaw.find_all('tr', attrs={'class', 'BdT Bdc($seperatorColor) Ta(end) Fz(s) Whs(nw)'})
    for result in results:
        rowValue = []
        for val in result.find_all('span'):
            rowValue.append(val.text)
            rowValue[0] = pd.to_datetime(rowValue[0])
        stockStats.append(rowValue)
    
    listOfStocks.append(stockStats)

print(listOfStocks)
```
